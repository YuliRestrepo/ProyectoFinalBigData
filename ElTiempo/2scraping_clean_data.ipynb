```python
import pandas as pd
import numpy as np
import json
import sys
from bs4 import BeautifulSoup
import requests
```

```python
texto=""
f=open("noticias_tecnologia.txt","r",encoding='utf-8')
texto=f.read()
f.close()
```

```python
print(texto)
```

```python
import bleach
clean = bleach.clean(texto, tags=[], strip=True)
print(clean)
```

```python
clean1=clean.replace("[","").replace("]","").replace(",","").replace("(","").replace(")","")
print(clean1)
```

```python
with open('noticias_tecnologia.txt','w',encoding='utf-8-sig') as f:
    f.write(str(clean1)+'\n')
```

```python
import re
import string

def clean_text_round1(text):
    text=text.lower()
    text=re.sub('\[.*?¿\]\%', ' ', text)
    text=re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    text=re.sub('\w*\d\w*', '', text)
    return text

round1 = clean_text_round1(clean1)
print (round1)
```

```python
with open('noticias_tecnologia2.txt', 'w', encoding='utf-8-sig') as f:
    f.write(str(round1)+'\n')
```
