### Trigrama

```python
tfidf = TfidfVectorizer([texto],ngram_range = (3,3))
```

```python
tfidf
```

```python
response = tfidf.fit_transform([texto])
```

```python
feature_names = tfidf.get_feature_names()
trigrama=[]
tf=[]
for col in response.nonzero()[1]:
    print (feature_names[col], ' - ', response[0, col])
    trigrama.append(feature_names[col])
    tf.append(response[0, col])
```

```python
arr={'trigrama':trigrama, 'tfidf':tf}
datos=pd.DataFrame(data=arr)
datos
```

```python
mediana = datos["tfidf"].median()
mediana
```

```python
moda = datos["tfidf"].mode()
moda
```

```python
media = datos["tfidf"].mean()
media
```

```python
medmas=mediana+0.000033
medmen=mediana-0.000033
```
