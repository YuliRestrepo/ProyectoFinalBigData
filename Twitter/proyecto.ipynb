```python
import pandas as pd
f = open("texto.txt", encoding="utf8")
df=f.read()
```

```python
import re
import unicodedata
import nltk
nltk.download('wordnet')
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt
```

```python
def basic_clean(text):

    wnl = nltk.stem.WordNetLemmatizer()
    stopwords = nltk.corpus.stopwords.words('spanish')
    text = (unicodedata.normalize('NFKD', text)
      .encode('ascii', 'ignore')
      .decode('utf-8', 'ignore')
      .lower())
    words = re.sub(r'[^\w\s]', '', text).split()
    return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

```python
words = basic_clean(df)
```

```python
unigrams=nltk.ngrams(words,1)
unifrequency=Counter(unigrams)
valueofunigrams=list(unifrequency.values())
unigramslist=list(unifrequency)

for x in range(0,len(unigramslist)):
    if(valueofunigrams[x]>=2 and valueofunigrams[x]<=5):
        print(unigramslist[x],"is used",valueofunigrams[x],"times")
```

```python
bigrams=nltk.ngrams(words,2)
bigramsfrequency=Counter(bigrams)
valueofbigram=list(bigramsfrequency.values())
bigramslist=list(bigramsfrequency)

for x in range(0,len(bigramslist)):
    if(valueofbigram[x]>=2 and valueofbigram[x]<=5):
        print(bigramslist[x],"is used",valueofbigram[x],"times")
```

```python
threegrams = nltk.ngrams(words, 3)
threegramsfrequency = Counter(threegrams)
valuesofthreegrams = list(threegramsfrequency.values())
threegramlist = list(threegramsfrequency)

for x in range(0, len(threegramlist)-1):
    if(valuesofthreegrams[x]>=2 and valuesofthreegrams[x]<=5):
        print(threegramlist[x], " is used ", valuesofthreegrams[x], " times")
```

```python
unigrams_series= (pd.Series(nltk.ngrams(words,1)).value_counts())[:10]
bigrams_series = (pd.Series(nltk.ngrams(words, 2)).value_counts())[:10]
trigrams_series = (pd.Series(nltk.ngrams(words, 3)).value_counts())[:10]
```

```python
trigrams_series.sort_values().plot.barh(color='blue', width=.9, figsize=(12, 8))
plt.title('Diagrama de trigramas')
plt.ylabel('Trigramas')
plt.xlabel('numero de veces')
```

```python
bigrams_series.sort_values().plot.barh(color='blue', width=.9, figsize=(12, 8))
plt.title('Diagrama de bigrams')
plt.ylabel('Bigram')
plt.xlabel('numero de veces')
```

```python
unigrams_series.sort_values().plot.barh(color='blue', width=.9, figsize=(12, 8))
plt.title('Diagrama de Unigrams')
plt.ylabel('Unigramas')
plt.xlabel('numero de veces')
```
