```python
import pandas as pd
f = open("texto.txt", encoding="utf8")
df=f.read()
```

```python
import re
import unicodedata
import nltk
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer([df],ngram_range = (1,1))
```

```python
response = tfidf.fit_transform([df])
```

```python
feature_names = tfidf.get_feature_names()
trigramas=[]
tf=[]
for col in response.nonzero()[1]:
    print (feature_names[col], ' - ', response[0, col])
    trigramas.append(feature_names[col])
    tf.append(response[0, col])
```

```python
arr={'unigrams':trigramas,'tfidf':tf}
datos=pd.DataFrame(data=arr)
datos
```

```python
mediana = datos["tfidf"].median()
media = datos["tfidf"].mean()
moda = datos["tfidf"].mode()
moda
```

```python
medmas=media+0.00033
medmen=media-0.00033
```

```python
datos[(datos['tfidf']>=medmen) & (datos['tfidf']<=medmas)]
```
